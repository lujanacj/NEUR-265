```python
#2. Importing packages
import numpy as np

from matplotlib import pyplot as plt

import scipy.stats as stats
```

```python
#3. Importing data as numpy arrays
v_rest_sst_url = 'https://raw.githubusercontent.com/hallockh/neur_265/b398c90e30efcabf56f6d1122ee5d338ec21ce93/data/v_rest_sst.csv'
v_rest_pvalb_url= 'https://raw.githubusercontent.com/lujanacj/NEUR-265/main/v_rest_pvalb.csv'

v_rest_sst = np.loadtxt(v_rest_sst_url)
v_rest_pvalb = np.loadtxt(v_rest_pvalb_url)
```

```python
#4.
#a. Finding mean of data sets
np.mean(v_rest_sst)
np.mean(v_rest_pvalb)

#b. T-test
pvalue=stats.ttest_ind(v_rest_sst,v_rest_pvalb)[1]
print(pvalue)
  #With a pvalue= 7.16E-9 < 0.05, the null hypothesis is refuted, meaning that the difference in the two means is significant enough to imply that they from different distributions

#c. Plotting Historgrams
if (pvalue <= 0.05):
  fig, ax = plt.subplots(1,2)
  ax[0].hist(v_rest_sst)
  ax[0].set_title('v_rest_sst')
  ax[1].hist(v_rest_pvalb)
  ax[1].set_title('v_rest_pvalb')

  #d. Labeling axes
  ax[0].set_xlabel('Resting Membrane Potential (mV)')
  ax[1].set_xlabel('Resting Membrane Potential (mV)')

  ax[0].set_ylabel('Frequency')
  ax[1].set_ylabel('Frequency')
```

```python
#5. vrestHist() function
def vrestHist():
    """
    Plots histograms of two data sets if they have statistcally significant means
    """
    #a. Finding mean of data sets
    np.mean(v_rest_sst)
    np.mean(v_rest_pvalb)

    #b. T-test
    pvalue=stats.ttest_ind(v_rest_sst,v_rest_pvalb)[1]

    #c. Plotting Historgrams
    if (pvalue <= 0.05):
      fig, ax = plt.subplots(1,2)
      ax[0].hist(v_rest_sst)
      ax[0].set_title('v_rest_sst')
      ax[1].hist(v_rest_pvalb)
      ax[1].set_title('v_rest_pvalb')

      #d. Labeling axes
      ax[0].set_xlabel('Resting Membrane Potential (mV)')
      ax[1].set_xlabel('Resting Membrane Potential (mV)')

      ax[0].set_ylabel('Frequency')
      ax[1].set_ylabel('Frequency')
```

```python
#6. Calling vrestHist()
vrestHist()
```
