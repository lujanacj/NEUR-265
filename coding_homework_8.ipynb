```python
#2. Importing Modules

import numpy as np

import matplotlib.pyplot as plt

from scipy.stats import pearsonr

```

```python
#3. Importing Data

url_extinction = 'https://raw.githubusercontent.com/lujanacj/NEUR-265/main/extinction_predictors_hw8.csv'
url_fear = 'https://raw.githubusercontent.com/lujanacj/NEUR-265/main/fear_conditioning_traces_hw8.csv'
url_tone = 'https://raw.githubusercontent.com/lujanacj/NEUR-265/main/tone_index.csv'
url_shock = 'https://raw.githubusercontent.com/lujanacj/NEUR-265/main/shock_index.csv'

extinction_predictors= np.loadtxt(url_extinction, delimiter = ',')
fear_traces= np.loadtxt(url_fear, delimiter = ',')
tone_index= np.loadtxt(url_tone, delimiter = ',').astype(int)
shock_index= np.loadtxt(url_shock, delimiter = ',').astype(int)

```

```python
#5. Conditioning and Extinction Arrays

conditioning_traces = fear_traces[0:10458,:]
extinction_traces = fear_traces[10459:-1,:]
```

```python
#12. Predictor Conditoning Model

shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(14208)
for i in range(14208):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = fear_traces[0:10458,0]


predictors = df(data={'Intercept': np.ones_like(shock_array), 'shock onset': shock_array})


predictors['shock onset'] = shock_array
model3 = sm.GLM(first_trace, predictors, family=Poisson())
model3_results = model3.fit() # Fit model to our data
b1 = model3_results.params
print('b1:')
print(b1)
```

```python
#13. r_model_conditioning variable

predictors['shock onset'] = shock_array
r_model_conditioning = np.zeros(63)

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r_index_conditioning = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r_index_conditioning.statistic
```

```python
#14. r_model_extinction variable

extinction_predictors = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
extinction_predictors['tone onset'] = tone_array
r_model_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  model3 = sm.GLM(trace_temp, extinction_predictors, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r_index_extinction = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r_index_extinction.statistic
```

```python
#15. Conditioning and Extinction index corelation

pearsonr(r_index_conditioning,r_index_extinction)
```

```python
#16. Scatter plot
plt.scatter(r_model_conditioning,r_model_extinction)
plt.xlabel('Conditioning')
plt.ylabel('Extinction')
```
