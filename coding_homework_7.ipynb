```python
#2. Importing Modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
#3. Importing Data
url_left1='https://raw.githubusercontent.com/lujanacj/NEUR-265/main/spike_left_trial1%20-%20Sheet1.csv' #for some reason it kept saying there was no data in the file so I copy and pasted it to a new file
url_left2='https://raw.githubusercontent.com/lujanacj/NEUR-265/main/spikes_left_trial2.csv'
url_right1='https://raw.githubusercontent.com/lujanacj/NEUR-265/main/spikes_right_trial1.csv'
url_right2='https://raw.githubusercontent.com/lujanacj/NEUR-265/main/spikes_right_trial2.csv'

left1=np.loadtxt(url_left1,delimiter=',')
left2=np.loadtxt(url_left2,delimiter=',')
right1=np.loadtxt(url_right1,delimiter=',')
right2=np.loadtxt(url_right2,delimiter=',')
```

```python
#4. Array Length
print(left1.size)
print(left2.size)
print(right1.size)
print(right2.size)
```

```python
#6. Firing Rates

period=30

rate_left1=left1.size/period
rate_left2=left2.size/period
rate_right1=right1.size/period
rate_right2=right2.size/period

print(rate_left1)
print(rate_left2)
print(rate_right1)
print(rate_right2)
```

#7. Firing Rate Interpretation
There appears to be no trend in firing rate with direction of turning, as the right trial has both the highest and lowest firing rates.

```python
#8. Spike Trains
fig, ax = plt.subplots(2)

plt.subplots_adjust(hspace=0.6)

ax[0].plot(left1, np.ones_like(left1), '.')
ax[0].plot(right1, 2 * np.ones_like(right1), '.')
ax[0].set_ylim([0, 3])
ax[0].set_xlabel('Time [s]')
ax[0].legend({'Left', 'Right'},loc='upper right')
ax[0].set_yticks([])
ax[0].set_title('Trial 1')

ax[1].plot(left2, np.ones_like(left2), '.')
ax[1].plot(right2, 2 * np.ones_like(right2), '.')
ax[1].set_ylim([0, 3])
ax[1].set_xlabel('Time [s]')
ax[1].legend({'Left', 'Right'},loc='upper right')
ax[1].set_yticks([])
ax[1].set_title('Trial 2')
```

#9. Spike Train Interpretation

Both the right and left turn trial spike trains consist of a decent amount of gaps in nueron firing, which appear to vary in both length and spacing. While the frequency of the gaps for the left turn is similar to that of the right's, the size of the gaps is generally much smaller, with the exception of one large gap at the end of trial 1.

```python
#10. ISI Arrays
ISI_left1=np.diff(left1)
ISI_left2=np.diff(left2)
ISI_right1=np.diff(right1)
ISI_right2=np.diff(right2)
```

```python
#11. ISI Histograms
fig, ax = plt.subplots(2,2)
plt.subplots_adjust(hspace=0.6)
bins = np.arange(0, 2, 0.01)

ax[0,0].hist(ISI_left1, bins)
ax[0,0].set_xlabel('ISI [s]')
ax[0,0].set_ylabel('Counts')
ax[0,0].set_title('Left Trial 1')

ax[1,0].hist(ISI_left2, bins)
ax[1,0].set_xlabel('ISI [s]')
ax[1,0].set_ylabel('Counts')
ax[1,0].set_title('Left Trial 2')

ax[0,1].hist(ISI_right1, bins)
ax[0,1].set_xlabel('ISI [s]')
ax[0,1].set_ylabel('Counts')
ax[0,1].set_title('Right Trial 1')

ax[1,1].hist(ISI_right2, bins)
ax[1,1].set_xlabel('ISI [s]')
ax[1,1].set_ylabel('Counts')
ax[1,1].set_title('Right Trial 2')
```

#12. Histogram Interpretation
The high concentration of low interspike intervals shows that these nuerons do fire in bursts. Again, there does not appear to be a drastic difference between the right and left trials, as both have relatively similar amounts of low interspike intervals.

```python
#13. Firing Rate Across Time
fig, ax = plt.subplots(2)
plt.subplots_adjust(hspace=0.6)

binsize = 1
fr_bins = np.arange(0,30,binsize)

fr_left1 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (left1 > fr_bins[i]) & (left1 < fr_bins[i+1])
  true_values = left1[spk_index]
  fr_left1[i] = (len(true_values))/binsize

  fr_left2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (left2 > fr_bins[i]) & (left2 < fr_bins[i+1])
  true_values = left2[spk_index]
  fr_left2[i] = (len(true_values))/binsize

  fr_right1 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (right1 > fr_bins[i]) & (right1 < fr_bins[i+1])
  true_values = right1[spk_index]
  fr_right1[i] = (len(true_values))/binsize

  fr_right2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_index = (right2 > fr_bins[i]) & (right2 < fr_bins[i+1])
  true_values = right2[spk_index]
  fr_right2[i] = (len(true_values))/binsize



ax[0].plot(fr_bins[:-1], fr_left1,color='red')
ax[0].plot(fr_bins[:-1], fr_right1,color='blue')
ax[1].plot(fr_bins[:-1], fr_left2,color='red')
ax[1].plot(fr_bins[:-1], fr_right2,color='blue')
plt.xlabel('Time [s]')
ax[0].set_ylabel('Firing Rate [Hz]')
ax[1].set_ylabel('Firing Rate [Hz]')
ax[0].set_title('Trial 1')
ax[1].set_title('Trial 2')
ax[0].legend({'Left', 'Right'},loc='upper right')
ax[1].legend({'Left', 'Right'},loc='upper right')
```

#14. Firing Rate Plot Interpretation
In trial 1, the firing rates start out relatively similar between the left and right turn trials, yet near the 10 second mark, the two diverge as the firing rates of the left decreases while the rates of the right drastically spikes up in magnitude. The second trial does not display this same pattern however, as the left starts up with high firing rates which continously decreases, while the right displays lower firing rates that remain relatively stangnant in magnitude.
