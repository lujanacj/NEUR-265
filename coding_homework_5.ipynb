```python
#2. Importing Modules
import numpy as np

import matplotlib.pyplot as plt
```

```python
#3. Importing Data
url_hits = 'https://raw.githubusercontent.com/lujanacj/NEUR-265/main/acc_hits_hw5.csv'
url_fa = 'https://raw.githubusercontent.com/lujanacj/NEUR-265/main/acc_fa_hw5.csv'

acc_hits = np.loadtxt(url_hits, delimiter = ',')
acc_fa = np.loadtxt(url_fa, delimiter = ',')
```

```python
#4. Shape of Arrays
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#6. Time Vector
time = np.linspace(0,4,acc_fa.shape[1])
```

```python
#7. Sampling Rate
samplingRate = acc_fa.shape[1]/4
```

```python
#8. Plots of First Trial
fig, ax = plt.subplots(2)

ax[0].plot(time,acc_hits[0,:])
ax[1].plot(time,acc_fa[0,:])


plt.xlabel('Time (s)')
plt.ylabel('Voltage ($\mu V$)')
```

#9. Identifying the Dominant Frequency
Both data sets appear to have a dominant frequency of ~1 Hz.

In the hit data, a trough is located at 2.5 seconds with a peak occurring at 3 seconds, making 0.5 seconds as half the period length, thus 1 second as the full period length.

In the false alarm data, two sequential troughs are located at the 2 and 3 second marks, making the period length to be 1 second as well.

Frequency is the inverse of period length, thus a period of 1 second results in a frequency of 1 Hz.

```python
#10. ERP hit data
plt.plot(time,acc_hits[0])
plt.plot([2,2],[-160,160],color='black')
plt.plot([time[0],time[-1]],[0,0],color='black')

plt.xlabel('Time (s)')
plt.ylabel('Voltage ($\mu V$)')
```

#11. ERP Interpretation
There appears to be an action potential after the stimulus at the 2 second mark

```python
#12. ERP of hit and false alarm data
plt.plot(time,acc_hits[0])
plt.plot(time,acc_fa[0])
plt.plot([2,2],[-160,160],color='black')
plt.plot([time[0],time[-1]],[0,0],color='black')

plt.xlabel('Time (s)')
plt.ylabel('Voltage ($\mu V$)')
```
